```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '../data/train.json'
out_file = '../data/train_processed.csv'
```

```python
df = pd.read_json(file)
df
```

```python
#  Определим количество данных в каждом классе
df['sentiment'].value_counts()
```

```python
# Упс, выборка не сбалансированная. Придется убрать лишнее
```

```python
# Выберем отдельно позитивные, негативные и нейтральные новости
keys = df['sentiment'].unique()
data = {}
for key in keys:
    data[key] = df[df['sentiment'] == key]
```

```python
# Перемешаем
for key in keys:
    data[key] = data[key].sample(frac=1).reset_index(drop=True)
```

```python
# Обрежем по самому короткому
min_len = min(len(d) for d in data.values())
print('Min length:',min_len)

for key in keys:
    data[key] = data[key][:min_len]
```

```python
# Соберем обратно
df = pd.concat(data.values())
df
```

```python
df.to_csv(out_file, index=False)
```
